```python
import time
import warnings

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import (StratifiedKFold, train_test_split, cross_val_score, 
                                   GridSearchCV)
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix)
from sklearn.datasets import fetch_openml

warnings.filterwarnings('ignore')
```

```python
plt.style.use('default')
sns.set_palette("husl")

print("="*60)
print("MLP CLASSIFICATION ON MNIST DATASET")
print("="*60)

print("Loading MNIST dataset...")
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='auto')
X, y = mnist.data, mnist.target.astype(int)

print(f"Original dataset shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Classes: {np.unique(y)}")

# Use subset for analysis
SAMPLE_SIZE = 15000
print(f"Using subset of {SAMPLE_SIZE} samples for analysis...")

X_subset, _, y_subset, _ = train_test_split(
    X, y, train_size=SAMPLE_SIZE, random_state=42, stratify=y
)

print(f"Subset shape: {X_subset.shape}")
print(f"Pixel values range: [{X_subset.min()}, {X_subset.max()}]")


feature_names = [f'pixel_{i}' for i in range(X_subset.shape[1])]
df = pd.DataFrame(X_subset, columns=feature_names)
df['target'] = y_subset
df['digit'] = y_subset.astype(str)

print(f"\nDataset Statistics:")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Duplicated rows: {df.duplicated().sum()}")
print(f"Class distribution:\n{df['target'].value_counts().sort_index()}")


plt.figure(figsize=(15, 10))
target_counts = df['target'].value_counts().sort_index()
bars = plt.bar(range(len(target_counts)), target_counts.values, 
               color=plt.cm.tab10(np.linspace(0, 1, 10)), alpha=0.7)
plt.title('MNIST Digit Distribution')
plt.xlabel('Digit')
plt.ylabel('Count')
plt.xticks(range(len(target_counts)), target_counts.index)
plt.show()


sample_indices = []
for digit in range(10):
    digit_indices = np.where(y_subset == digit)[0]
    if len(digit_indices) > 0:
        sample_indices.append(digit_indices[0])

sample_images = X_subset[sample_indices]
fig_samples = plt.figure(figsize=(12, 3))
fig_samples.suptitle('Sample Digits from MNIST Dataset', fontsize=14)
for i in range(10):
    plt.subplot(1, 10, i+1)
    plt.imshow(sample_images[i].reshape(28, 28), cmap='gray')
    plt.title(f'Digit {i}')
    plt.axis('off')
plt.tight_layout()
plt.show()


plt.figure(figsize=(15, 5))
plt.hist(X_subset.flatten(), bins=50, alpha=0.7, color='skyblue')
plt.title('Pixel Intensity Distribution')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()

X_normalized = X_subset / 255.0
print(f"Data normalized to range: [{X_normalized.min():.3f}, {X_normalized.max():.3f}]")

X_train, X_test, y_train, y_test = train_test_split(
    X_normalized, y_subset, test_size=0.2, random_state=42, stratify=y_subset
)

print(f"\nTrain-Test Split:")
print(f"  Training set: {X_train.shape[0]} samples")
print(f"  Test set: {X_test.shape[0]} samples")
print(f"  Training target distribution: {np.bincount(y_train)}")
print(f"  Test target distribution: {np.bincount(y_test)}")

skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

models = {
    'MLP (100)': MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42),
    'MLP (100, 50)': MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=300, random_state=42),
    'MLP (128, 64)': MLPClassifier(hidden_layer_sizes=(128, 64), max_iter=300, random_state=42),
    'MLP (200, 100, 50)': MLPClassifier(hidden_layer_sizes=(200, 100, 50), max_iter=300, random_state=42),
    'MLP ReLU': MLPClassifier(hidden_layer_sizes=(100, 50), activation='relu', max_iter=300, random_state=42),
    'MLP Tanh': MLPClassifier(hidden_layer_sizes=(100, 50), activation='tanh', max_iter=300, random_state=42),
    'MLP Adam': MLPClassifier(hidden_layer_sizes=(100, 50), solver='adam', max_iter=300, random_state=42),
    'MLP L-BFGS': MLPClassifier(hidden_layer_sizes=(100, 50), solver='lbfgs', max_iter=300, random_state=42)
}



param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (100, 50), (128, 64), (150, 100, 50)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'lbfgs'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive']
}

# Use smaller subset for grid search due to computational constraints
X_train_small = X_train[:3000]
y_train_small = y_train[:3000]

grid_search = GridSearchCV(
    MLPClassifier(max_iter=300, random_state=42), 
    param_grid, 
    cv=3,
    scoring='accuracy', 
    n_jobs=-1,
    verbose=1
)

start_time = time.time()
grid_search.fit(X_train_small, y_train_small)
grid_time = time.time() - start_time

models['MLP (Best)'] = MLPClassifier(**grid_search.best_params_, max_iter=500, random_state=42)

print(f"Grid search completed in {grid_time:.2f} seconds")
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best CV score: {grid_search.best_score_:.4f}")

def evaluate_mlp_model(model, X_train, X_test, y_train, y_test, cv, model_name):
    print(f"\nTraining {model_name}...")
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time
    
    start_time = time.time()
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    predict_time = time.time() - start_time
    
    # Cross-validation (on smaller subset for speed)
    cv_start = time.time()
    cv_accuracy = cross_val_score(model, X_train[:3000], y_train[:3000], cv=cv, scoring='accuracy')
    cv_time = time.time() - cv_start
    
    train_acc = accuracy_score(y_train, y_pred_train)
    test_acc = accuracy_score(y_test, y_pred_test)

    report = classification_report(y_test, y_pred_test, output_dict=True)

    per_class_precision = [report[str(i)]['precision'] for i in range(10)]
    per_class_recall = [report[str(i)]['recall'] for i in range(10)]
    per_class_f1 = [report[str(i)]['f1-score'] for i in range(10)]
    
    metrics = {
        'Train Accuracy': train_acc,
        'Test Accuracy': test_acc,
        'Macro Avg Precision': report['macro avg']['precision'],
        'Macro Avg Recall': report['macro avg']['recall'],
        'Macro Avg F1-Score': report['macro avg']['f1-score'],
        'Weighted Avg F1-Score': report['weighted avg']['f1-score'],
        'CV Accuracy Mean': cv_accuracy.mean(),
        'CV Accuracy Std': cv_accuracy.std(),
        'Training Time': train_time,
        'Prediction Time': predict_time,
        'CV Time': cv_time,
        'Training Loss': model.loss_ if hasattr(model, 'loss_') else 0,
        'N Iterations': model.n_iter_ if hasattr(model, 'n_iter_') else 0
    }

    # Create comprehensive visualization
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Confusion Matrix
    ax1 = plt.subplot(3, 4, 1)
    cm = confusion_matrix(y_test, y_pred_test)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=range(10), yticklabels=range(10))
    plt.title(f'{model_name}\nConfusion Matrix')
    plt.xlabel('Predicted Digit')
    plt.ylabel('Actual Digit')
    
    # 2. Per-class Performance
    ax2 = plt.subplot(3, 4, 2)
    x = range(10)
    width = 0.25
    plt.bar([i - width for i in x], per_class_precision, width, label='Precision', alpha=0.7)
    plt.bar(x, per_class_recall, width, label='Recall', alpha=0.7)
    plt.bar([i + width for i in x], per_class_f1, width, label='F1-Score', alpha=0.7)
    plt.xlabel('Digit')
    plt.ylabel('Score')
    plt.title(f'{model_name}\nPer-Class Performance')
    plt.legend()
    plt.xticks(x)
    plt.grid(True, alpha=0.3)
    
    # 3. Cross-validation Scores
    ax3 = plt.subplot(3, 4, 3)
    plt.bar(range(len(cv_accuracy)), cv_accuracy, color='skyblue', alpha=0.7)
    plt.axhline(y=cv_accuracy.mean(), color='red', linestyle='--', 
                label=f'Mean: {cv_accuracy.mean():.3f}')
    plt.xlabel('Fold')
    plt.ylabel('Accuracy')
    plt.title(f'{model_name}\nCV Accuracy Scores')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 4. Performance Summary
    ax4 = plt.subplot(3, 4, 4)
    perf_metrics = ['Train Acc', 'Test Acc', 'Precision', 'Recall', 'F1-Score']
    perf_values = [train_acc, test_acc, metrics['Macro Avg Precision'], 
                   metrics['Macro Avg Recall'], metrics['Macro Avg F1-Score']]
    
    bars = plt.bar(range(len(perf_metrics)), perf_values, 
                   color=['lightblue', 'lightgreen', 'orange', 'pink', 'lightcoral'])
    plt.ylabel('Score')
    plt.title(f'{model_name}\nPerformance Summary')
    plt.xticks(range(len(perf_metrics)), perf_metrics, rotation=45)
    plt.ylim(0, 1.1)
    for i, v in enumerate(perf_values):
        plt.text(i, v + 0.02, f'{v:.3f}', ha='center', va='bottom', fontsize=8)
    plt.grid(True, alpha=0.3)
    
    # 5. Training Loss Curve (if available)
    ax5 = plt.subplot(3, 4, 5)
    if hasattr(model, 'loss_curve_'):
        plt.plot(model.loss_curve_, 'b-', linewidth=2)
        plt.xlabel('Iteration')
        plt.ylabel('Loss')
        plt.title(f'{model_name}\nTraining Loss Curve')
        plt.grid(True, alpha=0.3)
    else:
        plt.text(0.5, 0.5, 'Loss curve\nnot available', ha='center', va='center', 
                transform=ax5.transAxes, fontsize=12)
        plt.title(f'{model_name}\nTraining Loss Curve')
        plt.axis('off')
    
    # 6. Sample Predictions (Correct)
    ax6 = plt.subplot(3, 4, 6)
    correct_predictions = np.where(y_pred_test == y_test)[0][:5]
    if len(correct_predictions) > 0:
        sample_correct = X_test[correct_predictions[0]].reshape(28, 28)
        plt.imshow(sample_correct, cmap='gray')
        plt.title(f'Correct: True={y_test[correct_predictions[0]]}, Pred={y_pred_test[correct_predictions[0]]}')
        plt.axis('off')
    
    # 7. Sample Predictions (Incorrect)
    ax7 = plt.subplot(3, 4, 7)
    incorrect_predictions = np.where(y_pred_test != y_test)[0][:5]
    if len(incorrect_predictions) > 0:
        sample_incorrect = X_test[incorrect_predictions[0]].reshape(28, 28)
        plt.imshow(sample_incorrect, cmap='gray')
        plt.title(f'Incorrect: True={y_test[incorrect_predictions[0]]}, Pred={y_pred_test[incorrect_predictions[0]]}')
        plt.axis('off')
    
    # 8. Class Distribution in Test Set
    ax8 = plt.subplot(3, 4, 8)
    test_counts = np.bincount(y_test)
    plt.bar(range(len(test_counts)), test_counts, color='lightblue', alpha=0.7)
    plt.xlabel('Digit')
    plt.ylabel('Count')
    plt.title(f'{model_name}\nTest Set Distribution')
    plt.xticks(range(10))
    plt.grid(True, alpha=0.3)
    
    # 9. Prediction Distribution
    ax9 = plt.subplot(3, 4, 9)
    pred_counts = np.bincount(y_pred_test)
    plt.bar(range(len(pred_counts)), pred_counts, color='lightgreen', alpha=0.7)
    plt.xlabel('Digit')
    plt.ylabel('Count')
    plt.title(f'{model_name}\nPrediction Distribution')
    plt.xticks(range(10))
    plt.grid(True, alpha=0.3)
    
    # 10. Training Time Comparison
    ax10 = plt.subplot(3, 4, 10)
    time_metrics = ['Training', 'Prediction', 'CV']
    time_values = [train_time, predict_time, cv_time]
    plt.bar(time_metrics, time_values, color=['red', 'blue', 'green'], alpha=0.7)
    plt.ylabel('Time (seconds)')
    plt.title(f'{model_name}\nTime Performance')
    for i, v in enumerate(time_values):
        plt.text(i, v + max(time_values)*0.02, f'{v:.2f}s', ha='center', va='bottom')
    plt.grid(True, alpha=0.3)
    
    # 11. Per-Class Accuracy
    ax11 = plt.subplot(3, 4, 11)
    class_accuracies = []
    for digit in range(10):
        digit_mask = y_test == digit
        if np.sum(digit_mask) > 0:
            digit_acc = accuracy_score(y_test[digit_mask], y_pred_test[digit_mask])
            class_accuracies.append(digit_acc)
        else:
            class_accuracies.append(0)
    
    plt.bar(range(10), class_accuracies, color=plt.cm.tab10(np.linspace(0, 1, 10)), alpha=0.7)
    plt.xlabel('Digit')
    plt.ylabel('Accuracy')
    plt.title(f'{model_name}\nPer-Class Accuracy')
    plt.xticks(range(10))
    plt.ylim(0, 1.1)
    for i, v in enumerate(class_accuracies):
        plt.text(i, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=8)
    plt.grid(True, alpha=0.3)
    
    # 12. Model Architecture Info
    ax12 = plt.subplot(3, 4, 12)
    plt.text(0.1, 0.9, f"Model: {model_name}", fontsize=12, fontweight='bold')
    plt.text(0.1, 0.8, f"Hidden Layers: {model.hidden_layer_sizes}", fontsize=10)
    plt.text(0.1, 0.7, f"Activation: {model.activation}", fontsize=10)
    plt.text(0.1, 0.6, f"Solver: {model.solver}", fontsize=10)
    plt.text(0.1, 0.5, f"Alpha: {model.alpha}", fontsize=10)
    plt.text(0.1, 0.4, f"Test Accuracy: {test_acc:.4f}", fontsize=10)
    plt.text(0.1, 0.3, f"Training Time: {train_time:.2f}s", fontsize=10)
    plt.text(0.1, 0.2, f"Iterations: {metrics['N Iterations']}", fontsize=10)
    plt.text(0.1, 0.1, f"Final Loss: {metrics['Training Loss']:.4f}", fontsize=10)
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.axis('off')
    plt.title(f'{model_name}\nModel Information')
    
    plt.tight_layout()
    plt.show()
    
    return metrics, report


print("\n" + "="*50)
print("MODEL EVALUATION")
print("="*50)

results = {}
reports = {}

for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"EVALUATING {name.upper()}")
    print('='*50)
    
    metrics, report = evaluate_mlp_model(model, X_train, X_test, y_train, y_test, skf, name)
    results[name] = metrics
    reports[name] = report
    
    print(f"\nPerformance Metrics for {name}:")
    print("-" * 40)
    for metric_name, value in metrics.items():
        if 'Time' in metric_name:
            print(f"{metric_name:<20}: {value:.2f} seconds")
        elif metric_name in ['N Iterations', 'Training Loss']:
            if metric_name == 'N Iterations':
                print(f"{metric_name:<20}: {value}")
            else:
                print(f"{metric_name:<20}: {value:.4f}")
        else:
            print(f"{metric_name:<20}: {value:.4f}")

print("\n" + "="*60)
print("MODEL COMPARISON")
print("="*60)

comparison_df = pd.DataFrame(results).T
print(comparison_df.round(4))

metrics_to_compare = ['Test Accuracy', 'Macro Avg Precision', 'Macro Avg Recall', 'Macro Avg F1-Score']
fig, axes = plt.subplots(1, len(metrics_to_compare), figsize=(25, 5))
fig.suptitle('MLP Model Comparison - Key Metrics', fontsize=16)

colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'lightcoral']

for i, metric in enumerate(metrics_to_compare):
    values = [results[name][metric] for name in models.keys()]
    bars = axes[i].bar(range(len(models)), values, color=colors[i], alpha=0.7)
    
    axes[i].set_ylabel('Score')
    axes[i].set_title(f'{metric}')
    axes[i].set_xticks(range(len(models)))
    axes[i].set_xticklabels(models.keys(), rotation=45, ha='right')
    axes[i].grid(True, alpha=0.3)
    
    for j, (bar, v) in enumerate(zip(bars, values)):
        axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(values)*0.01, 
                    f'{v:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("BEST MODEL ANALYSIS")
print("="*60)

best_model_name = max(results.keys(), key=lambda x: results[x]['Test Accuracy'])
best_model = models[best_model_name]
best_metrics = results[best_model_name]

print(f"Best Model: {best_model_name}")
print(f"Test Accuracy: {best_metrics['Test Accuracy']:.4f}")
print(f"Cross-Validation Accuracy: {best_metrics['CV Accuracy Mean']:.4f} ± {best_metrics['CV Accuracy Std']:.4f}")
print(f"Training Time: {best_metrics['Training Time']:.2f} seconds")
print(f"Prediction Time: {best_metrics['Prediction Time']:.2f} seconds")
print(f"Number of Iterations: {best_metrics['N Iterations']}")
print(f"Final Training Loss: {best_metrics['Training Loss']:.4f}")


print("\n" + "="*50)
print("HYPERPARAMETER ANALYSIS")
print("="*50)

# Hidden layer size analysis
layer_configs = [(50,), (100,), (150,), (100, 50), (150, 100), (200, 100), (150, 100, 50)]
layer_accuracies = []
layer_times = []

print("Analyzing hidden layer configurations...")
for config in layer_configs:
    model_layer = MLPClassifier(hidden_layer_sizes=config, max_iter=300, random_state=42)
    start_time = time.time()
    scores = cross_val_score(model_layer, X_train[:3000], y_train[:3000], cv=3, scoring='accuracy')
    layer_time = time.time() - start_time
    layer_accuracies.append(scores.mean())
    layer_times.append(layer_time)
    print(f"Layers {config}: Accuracy={scores.mean():.4f} ± {scores.std():.4f}, Time={layer_time:.2f}s")


activations = ['logistic', 'tanh', 'relu']
activation_accuracies = []
activation_times = []

print("\nAnalyzing activation functions...")
for activation in activations:
    model_activation = MLPClassifier(hidden_layer_sizes=(100, 50), activation=activation, max_iter=300, random_state=42)
    start_time = time.time()
    scores = cross_val_score(model_activation, X_train[:3000], y_train[:3000], cv=3, scoring='accuracy')
    activation_time = time.time() - start_time
    activation_accuracies.append(scores.mean())
    activation_times.append(activation_time)
    print(f"Activation {activation}: Accuracy={scores.mean():.4f} ± {scores.std():.4f}, Time={activation_time:.2f}s")


alphas = [0.0001, 0.001, 0.01, 0.1, 1.0]
alpha_accuracies = []
alpha_times = []

print("\nAnalyzing regularization parameter (alpha)...")
for alpha in alphas:
    model_alpha = MLPClassifier(hidden_layer_sizes=(100, 50), alpha=alpha, max_iter=300, random_state=42)
    start_time = time.time()
    scores = cross_val_score(model_alpha, X_train[:3000], y_train[:3000], cv=3, scoring='accuracy')
    alpha_time = time.time() - start_time
    alpha_accuracies.append(scores.mean())
    alpha_times.append(alpha_time)
    print(f"Alpha {alpha}: Accuracy={scores.mean():.4f} ± {scores.std():.4f}, Time={alpha_time:.2f}s")


plt.figure(figsize=(20, 10))

plt.subplot(2, 4, 1)
config_labels = [str(config) for config in layer_configs]
plt.bar(range(len(layer_configs)), layer_accuracies, color='skyblue', alpha=0.7)
plt.xlabel('Hidden Layer Configuration')
plt.ylabel('Cross-Validation Accuracy')
plt.title('MLP Performance vs Hidden Layer Size')
plt.xticks(range(len(layer_configs)), config_labels, rotation=45, ha='right')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 2)
plt.bar(range(len(layer_configs)), layer_times, color='lightcoral', alpha=0.7)
plt.xlabel('Hidden Layer Configuration')
plt.ylabel('Training Time (seconds)')
plt.title('MLP Training Time vs Hidden Layer Size')
plt.xticks(range(len(layer_configs)), config_labels, rotation=45, ha='right')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 3)
plt.bar(range(len(activations)), activation_accuracies, color='lightgreen', alpha=0.7)
plt.xlabel('Activation Function')
plt.ylabel('Cross-Validation Accuracy')
plt.title('MLP Performance vs Activation Function')
plt.xticks(range(len(activations)), activations)
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 4)
plt.bar(range(len(activations)), activation_times, color='orange', alpha=0.7)
plt.xlabel('Activation Function')
plt.ylabel('Training Time (seconds)')
plt.title('MLP Training Time vs Activation Function')
plt.xticks(range(len(activations)), activations)
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 5)
plt.semilogx(alphas, alpha_accuracies, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Alpha (Regularization)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('MLP Performance vs Regularization')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 6)
plt.semilogx(alphas, alpha_times, 'ro-', linewidth=2, markersize=8)
plt.xlabel('Alpha (Regularization)')
plt.ylabel('Training Time (seconds)')
plt.title('MLP Training Time vs Regularization')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 7)
plt.scatter(layer_times, layer_accuracies, c=range(len(layer_configs)), cmap='viridis', s=100)
for i, config in enumerate(layer_configs):
    plt.annotate(str(config), (layer_times[i], layer_accuracies[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)
plt.xlabel('Training Time (seconds)')
plt.ylabel('Accuracy')
plt.title('MLP: Accuracy vs Training Time Trade-off')
plt.colorbar(label='Configuration Index')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 8)
if hasattr(best_model, 'loss_curve_'):
    plt.plot(best_model.loss_curve_, 'b-', linewidth=2, label='Training Loss')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.title(f'Learning Curve - {best_model_name}')
    plt.legend()
    plt.grid(True, alpha=0.3)
else:
    plt.text(0.5, 0.5, 'Learning curve\nnot available\nfor best model', 
            ha='center', va='center', fontsize=12)
    plt.title(f'Learning Curve - {best_model_name}')

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("SUMMARY AND CONCLUSIONS")
print("="*60)

print(f"""
MLP Analysis Summary for MNIST Dataset:

1. Dataset Characteristics:
   - Total samples used: {len(df)}
   - Features (pixels): {X_subset.shape[1]} (28x28 images)
   - Classes: 10 digits (0-9)
   - Training samples: {len(X_train)}
   - Test samples: {len(X_test)}

2. Best Performing Model: {best_model_name}
   - Test Accuracy: {best_metrics['Test Accuracy']:.4f}
   - Macro Average Precision: {best_metrics['Macro Avg Precision']:.4f}
   - Macro Average Recall: {best_metrics['Macro Avg Recall']:.4f}
   - Macro Average F1-Score: {best_metrics['Macro Avg F1-Score']:.4f}
   - Weighted Average F1-Score: {best_metrics['Weighted Avg F1-Score']:.4f}

3. Cross-Validation Results:
   - Mean Accuracy: {best_metrics['CV Accuracy Mean']:.4f} ± {best_metrics['CV Accuracy Std']:.4f}

4. Performance Characteristics:
   - Training Time: {best_metrics['Training Time']:.2f} seconds
   - Prediction Time: {best_metrics['Prediction Time']:.2f} seconds
   - Training Iterations: {best_metrics['N Iterations']}
   - Final Training Loss: {best_metrics['Training Loss']:.4f}
   - Best Hidden Layer Architecture: {best_model.hidden_layer_sizes}
   - Best Activation Function: {best_model.activation}
   - Best Solver: {best_model.solver}
   - Best Alpha (Regularization): {best_model.alpha}
""")
```
