```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import (StratifiedKFold, train_test_split, cross_val_score, 
                                   GridSearchCV)
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, 
                           roc_auc_score, roc_curve, precision_recall_curve)
from sklearn.datasets import load_breast_cancer
```

```python
plt.style.use('default')
sns.set_palette("husl")


cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
feature_names = cancer.feature_names
class_names = cancer.target_names

print(f"Original target values: {np.unique(y)}")
print(f"Target names: {class_names}")
print(f"Target mapping: 0 = {class_names[0]}, 1 = {class_names[1]}")


df = pd.DataFrame(X, columns=feature_names)
df['target'] = y.astype(int) 
df['diagnosis'] = [class_names[i] for i in y]  


print(f"Dataset shape: {df.shape}")
print(f"Number of features: {len(feature_names)}")
print(f"Classes: {class_names}")
print(f"Class distribution:\n{df['target'].value_counts().sort_index()}")
print(f"Class percentages:\n{(df['target'].value_counts(normalize=True).sort_index()*100).round(2)}")

print(f"\nDataset Statistics:")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Duplicated rows: {df.duplicated().sum()}")


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
target_counts = df['target'].value_counts().sort_index()
bars1 = plt.bar(range(len(target_counts)), target_counts.values, 
               color=['lightcoral', 'lightblue'], alpha=0.7)

plt.title('Target Distribution (Count)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.xticks(range(len(target_counts)), [class_names[i] for i in target_counts.index])

for i, v in enumerate(target_counts.values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom')

plt.subplot(1, 2, 2)
target_pct = df['target'].value_counts(normalize=True).sort_index()
bars2 = plt.bar(range(len(target_pct)), target_pct.values, 
               color=['lightcoral', 'lightblue'], alpha=0.7)

plt.title('Target Distribution (Percentage)')
plt.xlabel('Diagnosis')
plt.ylabel('Percentage')
plt.xticks(range(len(target_pct)), [class_names[i] for i in target_pct.index])

for i, v in enumerate(target_pct.values):
    plt.text(i, v + 0.01, f'{v:.1%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()


correlations = df.drop('diagnosis', axis=1).corr()['target'].abs().sort_values(ascending=False)
top_features = correlations[1:11].index.tolist()  # Top 10 features (excluding target)

print(f"\nTop 10 features most correlated with target:")
for i, feature in enumerate(top_features, 1):
    print(f"{i:2d}. {feature:<35} : {correlations[feature]:.4f}")


X_data = df[feature_names].values 
y_data = df['target'].values.astype(int)

print(f"Feature matrix shape: {X_data.shape}")
print(f"Target vector shape: {y_data.shape}")
print(f"Target data type: {y_data.dtype}")
print(f"Unique target values: {np.unique(y_data)}")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_data)

print(f"Data normalized:")
print(f"  Min value: {X_scaled.min():.3f}")
print(f"  Max value: {X_scaled.max():.3f}")
print(f"  Mean: {X_scaled.mean():.3f}")
print(f"  Std: {X_scaled.std():.3f}")


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_data, test_size=0.2, random_state=42, stratify=y_data
)

print(f"\nTrain-Test Split:")
print(f"  Training set: {X_train.shape[0]} samples")
print(f"  Test set: {X_test.shape[0]} samples")
print(f"  Training target distribution: {np.bincount(y_train)}")
print(f"  Test target distribution: {np.bincount(y_test)}")

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


models = {
    'AdaBoost (Default)': AdaBoostClassifier(random_state=42),
    'AdaBoost (50 estimators)': AdaBoostClassifier(
        n_estimators=50, 
        learning_rate=1.0,
        random_state=42
    ),
    'AdaBoost (100 estimators)': AdaBoostClassifier(
        n_estimators=100, 
        learning_rate=1.0,
        random_state=42
    ),
    'AdaBoost (Decision Stump)': AdaBoostClassifier(
        estimator=DecisionTreeClassifier(max_depth=1),
        n_estimators=50,
        learning_rate=1.0,
        random_state=42
    ),
}


param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.1, 0.5, 1.0, 2.0],
    'algorithm': ['SAMME', 'SAMME.R']
}


grid_search = GridSearchCV(
    AdaBoostClassifier(random_state=42), 
    param_grid, 
    cv=3,
    scoring='accuracy', 
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train, y_train)
models['AdaBoost (Best)'] = grid_search.best_estimator_

print(f"Best parameters: {grid_search.best_params_}")
print(f"Best CV score: {grid_search.best_score_:.4f}")


def evaluate_adaboost_model(model, X_train, X_test, y_train, y_test, cv, model_name):  
    model.fit(X_train, y_train)
    
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    cv_accuracy = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
    cv_f1 = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1')
    cv_roc_auc = cross_val_score(model, X_train, y_train, cv=cv, scoring='roc_auc')
    cv_precision = cross_val_score(model, X_train, y_train, cv=cv, scoring='precision')
    cv_recall = cross_val_score(model, X_train, y_train, cv=cv, scoring='recall')
    

    train_acc = accuracy_score(y_train, y_pred_train)
    test_acc = accuracy_score(y_test, y_pred_test)
    

    report = classification_report(y_test, y_pred_test, output_dict=True)
    

    metrics = {
        'Train Accuracy': train_acc,
        'Test Accuracy': test_acc,
        'Precision': report['1']['precision'],
        'Recall': report['1']['recall'],
        'F1-Score': report['1']['f1-score'],
        'ROC-AUC': roc_auc_score(y_test, y_pred_proba),
        'CV Accuracy Mean': cv_accuracy.mean(),
        'CV Accuracy Std': cv_accuracy.std(),
        'CV F1 Mean': cv_f1.mean(),
        'CV F1 Std': cv_f1.std(),
        'CV ROC-AUC Mean': cv_roc_auc.mean(),
        'CV ROC-AUC Std': cv_roc_auc.std(),
        'CV Precision Mean': cv_precision.mean(),
        'CV Recall Mean': cv_recall.mean()
    }
    

    fig = plt.figure(figsize=(20, 12))
    
    # 1. Confusion Matrix
    ax1 = plt.subplot(3, 4, 1)
    cm = confusion_matrix(y_test, y_pred_test)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, 
                yticklabels=class_names)
    plt.title(f'{model_name}\nConfusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # 2. ROC Curve
    ax2 = plt.subplot(3, 4, 2)
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc_score(y_test, y_pred_proba):.3f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name}\nROC Curve')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 3. Precision-Recall Curve
    ax3 = plt.subplot(3, 4, 3)
    precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_pred_proba)
    plt.plot(recall_curve, precision_curve)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title(f'{model_name}\nPrecision-Recall Curve')
    plt.grid(True, alpha=0.3)
    
    # 4. Cross-validation Accuracy Scores
    ax4 = plt.subplot(3, 4, 4)
    plt.bar(range(len(cv_accuracy)), cv_accuracy, color='skyblue', alpha=0.7)
    plt.axhline(y=cv_accuracy.mean(), color='red', linestyle='--', 
                label=f'Mean: {cv_accuracy.mean():.3f}')
    plt.xlabel('Fold')
    plt.ylabel('Accuracy')
    plt.title(f'{model_name}\nCV Accuracy Scores')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 5. Prediction Probability Distribution
    ax6 = plt.subplot(3, 4, 5)
    plt.hist(y_pred_proba[y_test == 0], bins=20, alpha=0.6, label='Malignant', color='red')
    plt.hist(y_pred_proba[y_test == 1], bins=20, alpha=0.6, label='Benign', color='blue')
    plt.xlabel('Predicted Probability (Benign)')
    plt.ylabel('Frequency')
    plt.title(f'{model_name}\nPrediction Distribution')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # 6. Cross-validation Metrics Comparison
    ax7 = plt.subplot(3, 4, 6)
    cv_metrics = ['Accuracy', 'F1', 'ROC-AUC', 'Precision', 'Recall']
    cv_values = [cv_accuracy.mean(), cv_f1.mean(), cv_roc_auc.mean(), 
                cv_precision.mean(), cv_recall.mean()]
    cv_errors = [cv_accuracy.std(), cv_f1.std(), cv_roc_auc.std(), 
                cv_precision.std(), cv_recall.std()]
    
    plt.bar(cv_metrics, cv_values, yerr=cv_errors, capsize=5, 
            color=['skyblue', 'lightgreen', 'orange', 'pink', 'lightcoral'])
    plt.ylabel('Score')
    plt.title(f'{model_name}\nCV Metrics (Mean ± Std)')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    
    # 7. Classification Report Heatmap
    ax11 = plt.subplot(3, 4, 7)
    report_df = pd.DataFrame({
        'Precision': [report['0']['precision'], report['1']['precision']],
        'Recall': [report['0']['recall'], report['1']['recall']],
        'F1-Score': [report['0']['f1-score'], report['1']['f1-score']]
    }, index=['Malignant', 'Benign'])
    
    sns.heatmap(report_df, annot=True, cmap='RdYlBu_r', fmt='.3f')
    plt.title(f'{model_name}\nClassification Metrics')
    
    # 8. Model Performance Summary
    ax12 = plt.subplot(3, 4, 8)
    perf_metrics = ['Train Acc', 'Test Acc', 'Precision', 'Recall', 'F1', 'ROC-AUC']
    perf_values = [train_acc, test_acc, metrics['Precision'], 
                   metrics['Recall'], metrics['F1-Score'], metrics['ROC-AUC']]
    
    bars = plt.bar(range(len(perf_metrics)), perf_values, 
                   color=['lightblue', 'lightgreen', 'orange', 'pink', 'lightcoral', 'yellow'])
    plt.ylabel('Score')
    plt.title(f'{model_name}\nPerformance Summary')
    plt.xticks(range(len(perf_metrics)), perf_metrics, rotation=45)
    plt.ylim(0, 1.1)
    for i, v in enumerate(perf_values):
        plt.text(i, v + 0.02, f'{v:.3f}', ha='center', va='bottom', fontsize=8)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return metrics, report


results = {}
reports = {}


for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"EVALUATING {name.upper()}")
    print('='*50)
    
    metrics, report = evaluate_adaboost_model(model, X_train, X_test, y_train, y_test, skf, name)
    results[name] = metrics
    reports[name] = report
    

    print(f"\nPerformance Metrics for {name}:")
    print("-" * 40)
    for metric_name, value in metrics.items():
        print(f"{metric_name:<20}: {value:.4f}")
    

print("\n" + "="*60)
print("MODEL COMPARISON")
print("="*60)

comparison_df = pd.DataFrame(results).T
print("\nComplete Model Comparison:")
print(comparison_df.round(4))


metrics_to_compare = ['Test Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
fig, axes = plt.subplots(1, len(metrics_to_compare), figsize=(25, 5))
fig.suptitle('AdaBoost Model Comparison - Key Metrics', fontsize=16)

colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'lightcoral']

for i, metric in enumerate(metrics_to_compare):
    values = [results[name][metric] for name in models.keys()]
    bars = axes[i].bar(range(len(models)), values, color=colors[i], alpha=0.7)
    axes[i].set_title(f'{metric}')
    axes[i].set_ylabel('Score')
    axes[i].set_xticks(range(len(models)))
    axes[i].set_xticklabels(models.keys(), rotation=45, ha='right')
    axes[i].grid(True, alpha=0.3)
    
    for j, (bar, v) in enumerate(zip(bars, values)):
        axes[i].text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005, 
                    f'{v:.3f}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()



print("\n" + "="*60)
print("BEST MODEL ANALYSIS")
print("="*60)


best_model_name = max(results.keys(), key=lambda x: results[x]['Test Accuracy'])
best_model = models[best_model_name]
best_metrics = results[best_model_name]

print(f"Best Model: {best_model_name}")
print(f"Test Accuracy: {best_metrics['Test Accuracy']:.4f}")
print(f"Cross-Validation Accuracy: {best_metrics['CV Accuracy Mean']:.4f} ± {best_metrics['CV Accuracy Std']:.4f}")


if hasattr(best_model, 'feature_importances_'):
    feature_importance = best_model.feature_importances_
    feature_importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': feature_importance
    }).sort_values('Importance', ascending=False)
    
    print(f"\nTop 15 Most Important Features ({best_model_name}):")
    print("-" * 60)
    for i, (_, row) in enumerate(feature_importance_df.head(15).iterrows(), 1):
        print(f"{i:2d}. {row['Feature']:<35} : {row['Importance']:.4f}")
    

    plt.figure(figsize=(12, 8))
    top_20_features = feature_importance_df.head(20)
    plt.barh(range(len(top_20_features)), top_20_features['Importance'])
    plt.yticks(range(len(top_20_features)), top_20_features['Feature'])
    plt.xlabel('Feature Importance')
    plt.title(f'{best_model_name} - Top 20 Feature Importances')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()


print("\n" + "="*60)
print("SUMMARY AND CONCLUSIONS")
print("="*60)

print(f"""
AdaBoost Analysis Summary for Breast Cancer Wisconsin Dataset:

1. Dataset Characteristics:
   - Total samples: {len(df)}
   - Features: {len(feature_names)}
   - Classes: {class_names[0]} ({np.sum(y_data==0)} samples), {class_names[1]} ({np.sum(y_data==1)} samples)
   - Class balance: {np.sum(y_data==1)/len(y_data)*100:.1f}% Benign, {np.sum(y_data==0)/len(y_data)*100:.1f}% Malignant

2. Best Performing Model: {best_model_name}
   - Test Accuracy: {best_metrics['Test Accuracy']:.4f}
   - Precision: {best_metrics['Precision']:.4f}
   - Recall: {best_metrics['Recall']:.4f}
   - F1-Score: {best_metrics['F1-Score']:.4f}
   - ROC-AUC: {best_metrics['ROC-AUC']:.4f}

3. Cross-Validation Results:
   - Mean Accuracy: {best_metrics['CV Accuracy Mean']:.4f} ± {best_metrics['CV Accuracy Std']:.4f}
   - Mean F1-Score: {best_metrics['CV F1 Mean']:.4f} ± {best_metrics['CV F1 Std']:.4f}
   - Mean ROC-AUC: {best_metrics['CV ROC-AUC Mean']:.4f} ± {best_metrics['CV ROC-AUC Std']:.4f}

""")

print("="*60)
print("ANALYSIS COMPLETE!")
print("="*60)
```
