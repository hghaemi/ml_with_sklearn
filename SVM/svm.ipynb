```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.svm import SVC
from sklearn.model_selection import (StratifiedKFold, train_test_split, cross_val_score, 
                                   GridSearchCV, validation_curve)
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix, 
                           roc_auc_score, roc_curve, precision_recall_curve)
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import time
from sklearn.datasets import load_iris
from sklearn.multiclass import OneVsRestClassifier
from sklearn.feature_selection import SelectKBest, f_classif
```

```python

plt.style.use('default')
sns.set_palette("husl")

print("Loading Iris dataset...")
iris_data = load_iris()
X, y = iris_data.data, iris_data.target

feature_names = iris_data.feature_names
class_names = iris_data.target_names

print(f"Dataset shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"Number of classes: {len(np.unique(y))}")
print(f"Classes: {class_names}")
print(f"Features: {feature_names}")

df = pd.DataFrame(X, columns=feature_names)
df['target'] = y
df['species'] = [class_names[i] for i in y]

print(f"\nDataset Statistics:")
print(f"Final dataset shape: {df.shape}")
print(f"Missing values: {df.isnull().sum().sum()}")
print(f"Duplicated rows: {df.duplicated().sum()}")
print(f"Class distribution:\n{df['target'].value_counts().sort_index()}")

plt.figure(figsize=(20, 15))

plt.subplot(3, 3, 1)
target_counts = df['target'].value_counts().sort_index()
bars = plt.bar(range(len(target_counts)), target_counts.values, 
               color=plt.cm.Set3(np.linspace(0, 1, len(target_counts))), alpha=0.8)
plt.title('Iris Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(range(len(target_counts)), [class_names[i] for i in target_counts.index], rotation=45)
for i, v in enumerate(target_counts.values):
    plt.text(i, v + max(target_counts.values)*0.01, str(v), ha='center', va='bottom', fontsize=10)

# Class distribution percentage
plt.subplot(3, 3, 2)
target_pct = df['target'].value_counts(normalize=True).sort_index()
plt.pie(target_pct.values, labels=[class_names[i] for i in target_pct.index], 
        autopct='%1.1f%%', colors=plt.cm.Set3(np.linspace(0, 1, len(target_pct))))
plt.title('Species Distribution (Percentage)')

# Feature correlation with target
plt.subplot(3, 3, 3)
feature_data = df.drop(['species'], axis=1)
correlations = feature_data.corr()['target'].abs().sort_values(ascending=False)[1:]  # Exclude target itself
plt.barh(range(len(correlations)), correlations.values)
plt.yticks(range(len(correlations)), [feat for feat in correlations.index])
plt.xlabel('Correlation with Target')
plt.title('Feature Correlations with Target')

# Feature distribution for sepal length
plt.subplot(3, 3, 4)
for class_idx in np.unique(y):
    class_data = df[df['target'] == class_idx]['sepal length (cm)']
    plt.hist(class_data, bins=10, alpha=0.6, label=class_names[class_idx])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.legend()

# Pairwise feature relationships (sepal length vs sepal width)
plt.subplot(3, 3, 5)
for class_idx in np.unique(y):
    class_mask = df['target'] == class_idx
    plt.scatter(df[class_mask]['sepal length (cm)'], df[class_mask]['sepal width (cm)'], 
               alpha=0.7, label=class_names[class_idx], s=50)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.grid(True, alpha=0.3)

# Pairwise feature relationships (petal length vs petal width)
plt.subplot(3, 3, 6)
for class_idx in np.unique(y):
    class_mask = df['target'] == class_idx
    plt.scatter(df[class_mask]['petal length (cm)'], df[class_mask]['petal width (cm)'], 
               alpha=0.7, label=class_names[class_idx], s=50)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width')
plt.legend()
plt.grid(True, alpha=0.3)

# Feature distribution for petal width
plt.subplot(3, 3, 7)
for class_idx in np.unique(y):
    class_data = df[df['target'] == class_idx]['petal width (cm)']
    plt.hist(class_data, bins=10, alpha=0.6, label=class_names[class_idx])
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Petal Width')
plt.legend()

# Feature importance based on variance
plt.subplot(3, 3, 8)
feature_vars = df[feature_names].var()
plt.bar(range(len(feature_vars)), feature_vars.values, color='lightcoral', alpha=0.7)
plt.ylabel('Variance')
plt.title('Feature Variance')
plt.xticks(range(len(feature_vars)), [f.split()[0] for f in feature_vars.index], rotation=45)
for i, v in enumerate(feature_vars.values):
    plt.text(i, v + max(feature_vars.values)*0.01, f'{v:.2f}', ha='center', va='bottom')

# Correlation heatmap
plt.subplot(3, 3, 9)
corr_matrix = df[feature_names].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.3f')
plt.title('Feature Correlation Heatmap')

plt.tight_layout()
plt.show()


X_data = df[feature_names].values
y_data = df['target'].values

print(f"Feature matrix shape: {X_data.shape}")
print(f"Target vector shape: {y_data.shape}")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_data)

print(f"Data standardized:")
print(f"  Mean: {X_scaled.mean():.3f}")
print(f"  Std: {X_scaled.std():.3f}")
print(f"  Min: {X_scaled.min():.3f}")
print(f"  Max: {X_scaled.max():.3f}")


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_data, test_size=0.3, random_state=42, stratify=y_data
)

print(f"\nTrain-Test Split:")
print(f"  Training set: {X_train.shape[0]} samples")
print(f"  Test set: {X_test.shape[0]} samples")
print(f"  Training target distribution: {np.bincount(y_train)}")
print(f"  Test target distribution: {np.bincount(y_test)}")


skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


models = {
    'SVM Linear': SVC(kernel='linear', random_state=42, probability=True),
    'SVM RBF (C=1)': SVC(kernel='rbf', C=1.0, random_state=42, probability=True),
    'SVM RBF (C=10)': SVC(kernel='rbf', C=10.0, random_state=42, probability=True),
    'SVM RBF (C=100)': SVC(kernel='rbf', C=100.0, random_state=42, probability=True),
    'SVM Polynomial (degree=2)': SVC(kernel='poly', degree=2, random_state=42, probability=True),
    'SVM Polynomial (degree=3)': SVC(kernel='poly', degree=3, random_state=42, probability=True),
    'SVM Sigmoid': SVC(kernel='sigmoid', random_state=42, probability=True),
    'SVM RBF (gamma=scale)': SVC(kernel='rbf', gamma='scale', random_state=42, probability=True),
    'SVM RBF (gamma=auto)': SVC(kernel='rbf', gamma='auto', random_state=42, probability=True),
}

param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
    'degree': [2, 3, 4]  # Only used for poly kernel
}

grid_search = GridSearchCV(
    SVC(random_state=42, probability=True), 
    param_grid, 
    cv=5,
    scoring='accuracy', 
    n_jobs=-1,
    verbose=1
)

start_time = time.time()
grid_search.fit(X_train, y_train)
grid_time = time.time() - start_time

models['SVM (Best)'] = SVC(**grid_search.best_params_, random_state=42, probability=True)

print(f"Grid search completed in {grid_time:.2f} seconds")
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best CV score: {grid_search.best_score_:.4f}")


def evaluate_svm_model(model, X_train, X_test, y_train, y_test, cv, model_name, class_names):
    print(f"\nTraining {model_name}...")
    start_time = time.time()
    model.fit(X_train, y_train)
    train_time = time.time() - start_time

    start_time = time.time()
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    predict_time = time.time() - start_time
    
    y_pred_proba = model.predict_proba(X_test)
    
    cv_start = time.time()
    cv_accuracy = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')
    try:
        cv_f1_macro = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1_macro')
    except:
        cv_f1_macro = cv_accuracy
    cv_time = time.time() - cv_start
    
    train_acc = accuracy_score(y_train, y_pred_train)
    test_acc = accuracy_score(y_test, y_pred_test)
    
    report = classification_report(y_test, y_pred_test, output_dict=True, zero_division=0)
    
    per_class_precision = []
    per_class_recall = []
    per_class_f1 = []
    
    for i in range(len(class_names)):
        if str(i) in report:
            per_class_precision.append(report[str(i)]['precision'])
            per_class_recall.append(report[str(i)]['recall'])
            per_class_f1.append(report[str(i)]['f1-score'])
        else:
            per_class_precision.append(0.0)
            per_class_recall.append(0.0)
            per_class_f1.append(0.0)
    
    try:
        roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr', average='macro')
    except:
        roc_auc = 0.0
    
    macro_avg = report.get('macro avg', {})
    weighted_avg = report.get('weighted avg', {})
    
    metrics = {
        'Train Accuracy': train_acc,
        'Test Accuracy': test_acc,
        'Macro Avg Precision': macro_avg.get('precision', 0.0),
        'Macro Avg Recall': macro_avg.get('recall', 0.0),
        'Macro Avg F1-Score': macro_avg.get('f1-score', 0.0),
        'Weighted Avg F1-Score': weighted_avg.get('f1-score', 0.0),
        'ROC-AUC (macro)': roc_auc,
        'CV Accuracy Mean': cv_accuracy.mean(),
        'CV Accuracy Std': cv_accuracy.std(),
        'CV F1 Macro Mean': cv_f1_macro.mean(),
        'CV F1 Macro Std': cv_f1_macro.std(),
        'Training Time': train_time,
        'Prediction Time': predict_time,
        'CV Time': cv_time,
        'Support Vectors': len(model.support_) if hasattr(model, 'support_') else 0
    }
    

    fig = plt.figure(figsize=(20, 12))
    
    # 1. Confusion Matrix
    ax1 = plt.subplot(3, 2, 1)
    cm = confusion_matrix(y_test, y_pred_test)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, 
                yticklabels=class_names)
    plt.title(f'{model_name}\nConfusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # 2. Per-class Performance
    ax2 = plt.subplot(3, 2, 2)
    x = range(len(class_names))
    width = 0.25
    plt.bar([i - width for i in x], per_class_precision, width, label='Precision', alpha=0.7)
    plt.bar(x, per_class_recall, width, label='Recall', alpha=0.7)
    plt.bar([i + width for i in x], per_class_f1, width, label='F1-Score', alpha=0.7)
    plt.xlabel('Species')
    plt.ylabel('Score')
    plt.title(f'{model_name}\nPer-Class Performance')
    plt.legend()
    plt.xticks(x, class_names, rotation=45)
    plt.grid(True, alpha=0.3)
    
    # 3. Cross-validation Scores
    ax3 = plt.subplot(3, 2, 3)
    cv_metrics = ['Accuracy', 'F1-Macro']
    cv_means = [cv_accuracy.mean(), cv_f1_macro.mean()]
    cv_stds = [cv_accuracy.std(), cv_f1_macro.std()]
    plt.bar(cv_metrics, cv_means, yerr=cv_stds, capsize=5, 
            color=['skyblue', 'lightgreen'], alpha=0.7)
    plt.ylabel('Score')
    plt.title(f'{model_name}\nCV Performance')
    plt.grid(True, alpha=0.3)
    for i, (mean, std) in enumerate(zip(cv_means, cv_stds)):
        plt.text(i, mean + std + 0.01, f'{mean:.3f}±{std:.3f}', ha='center', va='bottom', fontsize=9)
    
    # 4. Performance Summary
    ax4 = plt.subplot(3, 2, 4)
    perf_metrics = ['Train Acc', 'Test Acc', 'Precision', 'Recall', 'F1-Score']
    perf_values = [train_acc, test_acc, metrics['Macro Avg Precision'], 
                   metrics['Macro Avg Recall'], metrics['Macro Avg F1-Score']]
    
    bars = plt.bar(range(len(perf_metrics)), perf_values, 
                   color=['lightblue', 'lightgreen', 'orange', 'pink', 'lightcoral'])
    plt.ylabel('Score')
    plt.title(f'{model_name}\nPerformance Summary')
    plt.xticks(range(len(perf_metrics)), perf_metrics, rotation=45)
    plt.ylim(0, 1.1)
    for i, v in enumerate(perf_values):
        plt.text(i, v + 0.02, f'{v:.3f}', ha='center', va='bottom', fontsize=9)
    plt.grid(True, alpha=0.3)
    
    # 5. Class-wise Accuracy
    ax5 = plt.subplot(3, 2, 5)
    class_accuracies = []
    for class_idx in range(len(class_names)):
        class_mask = y_test == class_idx
        if np.sum(class_mask) > 0:
            class_acc = accuracy_score(y_test[class_mask], y_pred_test[class_mask])
            class_accuracies.append(class_acc)
        else:
            class_accuracies.append(0)
    
    plt.bar(range(len(class_names)), class_accuracies, 
            color=plt.cm.Set3(np.linspace(0, 1, len(class_names))), alpha=0.7)
    plt.xlabel('Species')
    plt.ylabel('Accuracy')
    plt.title(f'{model_name}\nPer-Class Accuracy')
    plt.xticks(range(len(class_names)), class_names, rotation=45)
    plt.ylim(0, 1.1)
    for i, v in enumerate(class_accuracies):
        plt.text(i, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontsize=9)
    plt.grid(True, alpha=0.3)
    
    # 6. Decision Boundary Visualization
    ax7 = plt.subplot(3, 2, 6)
    if X_train.shape[1] >= 2:

        pca_vis = PCA(n_components=2)
        X_train_pca = pca_vis.fit_transform(X_train)
        X_test_pca = pca_vis.transform(X_test)
        
        model_2d = SVC(**model.get_params())
        model_2d.fit(X_train_pca, y_train)
        
        h = 0.02
        x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
        y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                           np.arange(y_min, y_max, h))
        
        Z = model_2d.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        
        plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdYlBu)
        
        for class_idx in range(len(class_names)):
            class_mask = y_test == class_idx
            plt.scatter(X_test_pca[class_mask, 0], X_test_pca[class_mask, 1], 
                       c=plt.cm.RdYlBu(class_idx / (len(class_names)-1)), 
                       edgecolors='black', s=50, label=class_names[class_idx])
        
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.title(f'{model_name}\nDecision Boundary (PCA)')
        plt.legend()
    plt.tight_layout()
    plt.show()
    
    return metrics, report


results = {}
reports = {}

for name, model in models.items():
    print(f"\n{'='*50}")
    print(f"EVALUATING {name.upper()}")
    print('='*50)
    
    metrics, report = evaluate_svm_model(model, X_train, X_test, y_train, y_test, skf, name, class_names)
    results[name] = metrics
    reports[name] = report
    
    print(f"\nPerformance Metrics for {name}:")
    print("-" * 40)
    for metric_name, value in metrics.items():
        if 'Time' in metric_name:
            print(f"{metric_name:<25}: {value:.4f} seconds")
        else:
            print(f"{metric_name:<25}: {value:.4f}")


comparison_df = pd.DataFrame(results).T

metrics_to_compare = ['Test Accuracy', 'Macro Avg Precision', 'Macro Avg Recall', 'Macro Avg F1-Score']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('SVM Model Comparison - Key Metrics', fontsize=16)

colors = ['skyblue', 'lightgreen', 'orange', 'pink', 'lightcoral']

for i, metric in enumerate(metrics_to_compare):
    row = i // 3
    col = i % 3
    ax = axes[row, col]
    
    values = [results[name][metric] for name in models.keys()]
    bars = ax.bar(range(len(models)), values, color=colors[i], alpha=0.7)

    ax.set_ylabel('Score')
    format_str = '{:.3f}'

    ax.set_title(f'{metric}')
    ax.set_xticks(range(len(models)))
    ax.set_xticklabels(models.keys(), rotation=45, ha='right')
    ax.grid(True, alpha=0.3)
    
    for j, (bar, v) in enumerate(zip(bars, values)):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(values)*0.01, 
               format_str.format(v), ha='center', va='bottom', fontsize=8)


ax = axes[1, 2]
roc_values = [results[name]['ROC-AUC (macro)'] for name in models.keys()]
bars = ax.bar(range(len(models)), roc_values, color='purple', alpha=0.7)
ax.set_ylabel('ROC-AUC Score')
ax.set_title('ROC-AUC (macro)')
ax.set_xticks(range(len(models)))
ax.set_xticklabels(models.keys(), rotation=45, ha='right')
ax.grid(True, alpha=0.3)

for j, (bar, v) in enumerate(zip(bars, roc_values)):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + max(roc_values)*0.01, 
           f'{v:.3f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()


print("\n" + "="*50)
print("KERNEL COMPARISON ANALYSIS")
print("="*50)

kernel_results = {}
for name, model in models.items():
    if hasattr(model, 'kernel'):
        kernel = model.kernel
        if kernel not in kernel_results:
            kernel_results[kernel] = []
        kernel_results[kernel].append({
            'name': name,
            'accuracy': results[name]['Test Accuracy'],
            'f1': results[name]['Macro Avg F1-Score'],
            'training_time': results[name]['Training Time'],
            'roc_auc': results[name]['ROC-AUC (macro)']
        })


plt.figure(figsize=(20, 12))

plt.subplot(2, 4, 1)
kernel_names = list(kernel_results.keys())
kernel_accuracies = [np.mean([model['accuracy'] for model in kernel_results[k]]) for k in kernel_names]
bars = plt.bar(kernel_names, kernel_accuracies, color=plt.cm.viridis(np.linspace(0, 1, len(kernel_names))), alpha=0.7)
plt.ylabel('Average Test Accuracy')
plt.title('Average Accuracy by Kernel')
plt.xticks(rotation=45)
for i, v in enumerate(kernel_accuracies):
    plt.text(i, v + 0.005, f'{v:.3f}', ha='center', va='bottom')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 2)
kernel_f1_scores = [np.mean([model['f1'] for model in kernel_results[k]]) for k in kernel_names]
bars = plt.bar(kernel_names, kernel_f1_scores, color=plt.cm.plasma(np.linspace(0, 1, len(kernel_names))), alpha=0.7)
plt.ylabel('Average F1-Score')
plt.title('Average F1-Score by Kernel')
plt.xticks(rotation=45)
for i, v in enumerate(kernel_f1_scores):
    plt.text(i, v + 0.005, f'{v:.3f}', ha='center', va='bottom')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 3)
kernel_times = [np.mean([model['training_time'] for model in kernel_results[k]]) for k in kernel_names]
bars = plt.bar(kernel_names, kernel_times, color=plt.cm.coolwarm(np.linspace(0, 1, len(kernel_names))), alpha=0.7)
plt.ylabel('Average Training Time (seconds)')
plt.title('Average Training Time by Kernel')
plt.xticks(rotation=45)
for i, v in enumerate(kernel_times):
    plt.text(i, v + max(kernel_times)*0.02, f'{v:.4f}s', ha='center', va='bottom')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 4)
kernel_roc = [np.mean([model['roc_auc'] for model in kernel_results[k]]) for k in kernel_names]
bars = plt.bar(kernel_names, kernel_roc, color=plt.cm.spring(np.linspace(0, 1, len(kernel_names))), alpha=0.7)
plt.ylabel('Average ROC-AUC')
plt.title('Average ROC-AUC by Kernel')
plt.xticks(rotation=45)
for i, v in enumerate(kernel_roc):
    plt.text(i, v + 0.005, f'{v:.3f}', ha='center', va='bottom')
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 5)
all_accuracies = []
all_times = []
all_labels = []
colors_scatter = []
for i, (kernel, models_list) in enumerate(kernel_results.items()):
    for model in models_list:
        all_accuracies.append(model['accuracy'])
        all_times.append(model['training_time'])
        all_labels.append(model['name'])
        colors_scatter.append(plt.cm.tab10(i))

scatter = plt.scatter(all_times, all_accuracies, c=colors_scatter, s=100, alpha=0.7)
plt.xlabel('Training Time (seconds)')
plt.ylabel('Test Accuracy')
plt.title('Accuracy vs Training Time Trade-off')
plt.grid(True, alpha=0.3)


legend_elements = [plt.Line2D([0], [0], marker='o', color='w', 
                             markerfacecolor=plt.cm.tab10(i), markersize=10, label=kernel)
                  for i, kernel in enumerate(kernel_names)]
plt.legend(handles=legend_elements, title='Kernel Type', loc='best')


plt.subplot(2, 4, 6)
model_names_short = [name[:15] for name in models.keys()]
support_vectors = [results[name]['Support Vectors'] for name in models.keys()]
bars = plt.bar(range(len(model_names_short)), support_vectors, color='lightcoral', alpha=0.7)
plt.ylabel('Number of Support Vectors')
plt.title('Support Vectors by Model')
plt.xticks(range(len(model_names_short)), model_names_short, rotation=45)
for i, v in enumerate(support_vectors):
    plt.text(i, v + max(support_vectors)*0.02, str(v), ha='center', va='bottom', fontsize=8)
plt.grid(True, alpha=0.3)

plt.subplot(2, 4, 7)
cv_means = [results[name]['CV Accuracy Mean'] for name in models.keys()]
cv_stds = [results[name]['CV Accuracy Std'] for name in models.keys()]
plt.errorbar(range(len(model_names_short)), cv_means, yerr=cv_stds, fmt='o-', capsize=5)
plt.ylabel('CV Accuracy')
plt.title('Cross-Validation Stability')
plt.xticks(range(len(model_names_short)), model_names_short, rotation=45)
plt.grid(True, alpha=0.3)


plt.subplot(2, 4, 8)

complexity_scores = []
for name, model in models.items():
    complexity = 1
    if model.kernel == 'poly':
        complexity = 2 + (getattr(model, 'degree', 3) - 2) * 0.5
    elif model.kernel == 'rbf':
        complexity = 3
    elif model.kernel == 'sigmoid':
        complexity = 3.5
    
    if hasattr(model, 'C') and model.C > 10:
        complexity += 0.5
    
    complexity_scores.append(complexity)

accuracies_for_complexity = [results[name]['Test Accuracy'] for name in models.keys()]
plt.scatter(complexity_scores, accuracies_for_complexity, s=100, alpha=0.7, 
           c=range(len(models)), cmap='viridis')
plt.xlabel('Model Complexity Score')
plt.ylabel('Test Accuracy')
plt.title('Performance vs Complexity')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


print("\n" + "="*60)
print("BEST MODEL ANALYSIS")
print("="*60)

best_model_name = max(results.keys(), key=lambda x: results[x]['Test Accuracy'])
best_model = models[best_model_name]
best_metrics = results[best_model_name]

print(f"Best Model: {best_model_name}")
print(f"Test Accuracy: {best_metrics['Test Accuracy']:.4f}")
print(f"Cross-Validation Accuracy: {best_metrics['CV Accuracy Mean']:.4f} ± {best_metrics['CV Accuracy Std']:.4f}")
print(f"Training Time: {best_metrics['Training Time']:.4f} seconds")
print(f"Prediction Time: {best_metrics['Prediction Time']:.4f} seconds")
print(f"Number of Support Vectors: {best_metrics['Support Vectors']}")
print(f"ROC-AUC Score: {best_metrics['ROC-AUC (macro)']:.4f}")


print("\n" + "="*50)
print("HYPERPARAMETER ANALYSIS")
print("="*50)


C_values = [0.1, 1, 10, 100, 1000]
C_accuracies = []
C_times = []

print("Analyzing effect of C parameter (RBF kernel)...")
for C in C_values:
    model_C = SVC(kernel='rbf', C=C, random_state=42)
    start_time = time.time()
    scores = cross_val_score(model_C, X_train, y_train, cv=5, scoring='accuracy')
    C_time = time.time() - start_time
    C_accuracies.append(scores.mean())
    C_times.append(C_time)
    print(f"C={C:6.1f}: Accuracy={scores.mean():.4f} ± {scores.std():.4f}, Time={C_time:.3f}s")


gamma_values = [0.001, 0.01, 0.1, 1, 10]
gamma_accuracies = []
gamma_times = []

print("\nAnalyzing effect of gamma parameter (RBF kernel, C=1)...")
for gamma in gamma_values:
    model_gamma = SVC(kernel='rbf', C=1, gamma=gamma, random_state=42)
    start_time = time.time()
    scores = cross_val_score(model_gamma, X_train, y_train, cv=5, scoring='accuracy')
    gamma_time = time.time() - start_time
    gamma_accuracies.append(scores.mean())
    gamma_times.append(gamma_time)
    print(f"gamma={gamma:6.3f}: Accuracy={scores.mean():.4f} ± {scores.std():.4f}, Time={gamma_time:.3f}s")


plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.semilogx(C_values, C_accuracies, 'bo-', linewidth=2, markersize=8)
plt.xlabel('C Value (log scale)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('SVM Performance vs C Parameter')
plt.grid(True, alpha=0.3)
for i, (c, acc) in enumerate(zip(C_values, C_accuracies)):
    plt.annotate(f'{acc:.3f}', (c, acc), textcoords="offset points", xytext=(0,10), ha='center')

plt.subplot(1, 4, 2)
plt.semilogx(gamma_values, gamma_accuracies, 'ro-', linewidth=2, markersize=8)
plt.xlabel('Gamma Value (log scale)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('SVM Performance vs Gamma Parameter')
plt.grid(True, alpha=0.3)
for i, (g, acc) in enumerate(zip(gamma_values, gamma_accuracies)):
    plt.annotate(f'{acc:.3f}', (g, acc), textcoords="offset points", xytext=(0,10), ha='center')

plt.subplot(1, 4, 3)
plt.semilogx(C_values, C_times, 'go-', linewidth=2, markersize=8)
plt.xlabel('C Value (log scale)')
plt.ylabel('Training Time (seconds)')
plt.title('SVM Training Time vs C Parameter')
plt.grid(True, alpha=0.3)

plt.subplot(1, 4, 4)
plt.semilogx(gamma_values, gamma_times, 'mo-', linewidth=2, markersize=8)
plt.xlabel('Gamma Value (log scale)')
plt.ylabel('Training Time (seconds)')
plt.title('SVM Training Time vs Gamma Parameter')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


linear_models = [name for name in models.keys() if 'Linear' in name]
if linear_models:
    print("\n" + "="*50)
    print("FEATURE ANALYSIS (LINEAR SVM)")
    print("="*50)
    
    linear_model = models[linear_models[0]]
    linear_model.fit(X_train, y_train)
    
    if hasattr(linear_model, 'coef_'):
        if linear_model.coef_.ndim == 2:
            feature_importance_per_class = linear_model.coef_
            
            print("Feature Coefficients by Class (Linear SVM):")
            print("-" * 80)
            print(f"{'Feature':<25} {'Setosa':<12} {'Versicolor':<12} {'Virginica':<12}")
            print("-" * 80)
            
            for i, feature in enumerate(feature_names):
                coefs = [feature_importance_per_class[j, i] for j in range(len(class_names))]
                print(f"{feature:<25} {coefs[0]:>11.4f} {coefs[1]:>11.4f} {coefs[2]:>11.4f}")
            
            feature_importance = np.mean(np.abs(feature_importance_per_class), axis=0)
        else:
            feature_importance = np.abs(linear_model.coef_.flatten())
        
        print(f"\nOverall Feature Importance (Linear SVM):")
        print("-" * 50)
        feature_importance_pairs = list(zip(feature_names, feature_importance))
        feature_importance_pairs.sort(key=lambda x: x[1], reverse=True)
        
        for i, (name, importance) in enumerate(feature_importance_pairs, 1):
            print(f"{i}. {name:<30} : {importance:.4f}")
        

print("\n" + "="*50)
print("PERFORMANCE BY SPECIES")
print("="*50)

best_report = reports[best_model_name]
print(f"Detailed Performance Analysis ({best_model_name}):")
print("-" * 80)

for i, species in enumerate(class_names):
    if str(i) in best_report:
        precision = best_report[str(i)]['precision']
        recall = best_report[str(i)]['recall']
        f1_score = best_report[str(i)]['f1-score']
        support = best_report[str(i)]['support']
        
        print(f"{species:<15}: Precision={precision:.3f}, Recall={recall:.3f}, "
              f"F1={f1_score:.3f}, Support={support}")


print("\n" + "="*60)
print("SUMMARY AND CONCLUSIONS")
print("="*60)

print(f"""
SVM Analysis Summary for Iris Dataset:

1. Dataset Characteristics:
   - Total samples: {len(df)}
   - Features: {X_data.shape[1]} (sepal length, sepal width, petal length, petal width)
   - Classes: {len(class_names)} species (Setosa, Versicolor, Virginica)
   - Training samples: {len(X_train)}
   - Test samples: {len(X_test)}
   - Perfectly balanced dataset: 50 samples per class

2. Best Performing Model: {best_model_name}
   - Test Accuracy: {best_metrics['Test Accuracy']:.4f}
   - Macro Average Precision: {best_metrics['Macro Avg Precision']:.4f}
   - Macro Average Recall: {best_metrics['Macro Avg Recall']:.4f}
   - Macro Average F1-Score: {best_metrics['Macro Avg F1-Score']:.4f}
   - Weighted Average F1-Score: {best_metrics['Weighted Avg F1-Score']:.4f}
   - ROC-AUC (macro): {best_metrics['ROC-AUC (macro)']:.4f}

3. Cross-Validation Results:
   - Mean Accuracy: {best_metrics['CV Accuracy Mean']:.4f} ± {best_metrics['CV Accuracy Std']:.4f}
   - Mean F1-Score (macro): {best_metrics['CV F1 Macro Mean']:.4f} ± {best_metrics['CV F1 Macro Std']:.4f}

4. Best Model Characteristics:
   - Kernel: {best_model.kernel}
   - C parameter: {best_model.C if hasattr(best_model, 'C') else 'N/A'}
   - Gamma: {best_model.gamma if hasattr(best_model, 'gamma') else 'N/A'}
   - Support Vectors: {best_metrics['Support Vectors']}
   - Training Time: {best_metrics['Training Time']:.4f} seconds
   - Prediction Time: {best_metrics['Prediction Time']:.4f} seconds

5. Kernel Performance Ranking:""")


kernel_performance = {}
for name, model in models.items():
    if hasattr(model, 'kernel') and name in results:
        kernel = model.kernel
        if kernel not in kernel_performance:
            kernel_performance[kernel] = []
        kernel_performance[kernel].append(results[name]['Test Accuracy'])

for i, (kernel, accuracies) in enumerate(sorted(kernel_performance.items(), 
                                               key=lambda x: np.mean(x[1]), reverse=True), 1):
    avg_acc = np.mean(accuracies)
    print(f"   {i}. {kernel:<12}: Average Accuracy = {avg_acc:.4f}")

print(f"""

6. Key Observations for Iris Dataset:
   - Iris is a linearly separable dataset for most classes
   - Linear SVM performs excellently, often achieving perfect classification
   - RBF kernel can achieve near-perfect performance but may overfit
   - Setosa is easily separable from other classes
   - Versicolor and Virginica are more challenging to distinguish
   - Petal measurements are more discriminative than sepal measurements

7. Hyperparameter Insights:
   - Optimal C value: {grid_search.best_params_.get('C', 'N/A')}
   - Optimal gamma: {grid_search.best_params_.get('gamma', 'N/A')}
   - Grid search best accuracy: {grid_search.best_score_:.4f}
   - Lower C values work well due to linearly separable nature

8. Computational Considerations:
   - Very fast training due to small dataset size
   - Linear kernel is fastest and most interpretable
   - RBF kernel adds minimal computational cost for this size
   - Support vector count is low due to good separability

9. Feature Analysis (Linear SVM):
   - Petal length and petal width are most discriminative
   - Sepal width has lower discriminative power
   - Different features are important for different class pairs

10. Recommendations:
    - Linear SVM is ideal for Iris dataset due to linear separability
    - For similar small, well-separated datasets, start with linear kernel
    - Cross-validation is crucial even for small datasets
    - Feature scaling is important for SVM performance
    - Consider feature selection for higher-dimensional versions

""")

print("="*60)
print("ANALYSIS COMPLETE!")
print("="*60)
```
